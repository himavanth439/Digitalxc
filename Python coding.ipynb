```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = r"C:\Users\excel\OneDrive\Desktop\Sample Data file for Analysis_Jan'25.xlsx"
data = pd.read_excel(data)
data
```

```python
print("Dataset Overview:")
print(data.head())
```

```python
data = data.drop_duplicates()
```

```python
data = data.fillna({'Priority': 'Unknown', 'Assigned Group': 'Unknown', 'Technician': 'Unknown'})
```

```python
print(data.columns)
```

```python
print("Columns in dataset:")
print(data.columns)
```

```python
# Normalize column names
data.columns = data.columns.str.strip().str.lower().str.replace(" ", "_")
print("Normalized Columns:", data.columns)
```

```python
# Standardize date formats
data['created_date'] = pd.to_datetime(data['created_date'], errors='coerce')
data['resolved_date'] = pd.to_datetime(data['resolved_date'], errors='coerce')

# Handle invalid dates
invalid_created_dates = data[data['created_date'].isna()]
if not invalid_created_dates.empty:
    print("Rows with invalid 'created_date':")
    print(invalid_created_dates)

# Drop rows with missing created_date
data = data.dropna(subset=['created_date'])
```

```python
# Extract year, month, and day from created_date
data['year'] = data['created_date'].dt.year
data['month'] = data['created_date'].dt.month
data['day'] = data['created_date'].dt.day
```

```python
data
```

```python
# Calculate the average resolution time per category and priority
avg_resolution_time = data.groupby(['category', 'priority'])['resolution_time'].mean().reset_index()
avg_resolution_time.rename(columns={'resolution_time': 'avg_resolution_time'}, inplace=True)
```

```python
# Add a "closed" indicator based on the status column
data['closed'] = data['status'].str.lower().apply(lambda x: 1 if x == 'closed' else 0)

# Calculate closure rate by assigned group
closure_rate = data.groupby('assigned_group')['closed'].mean().reset_index()
closure_rate.rename(columns={'closed': 'closure_rate'}, inplace=True)
```

```python
# Aggregate metrics by year and month
monthly_summary = data.groupby([data['created_date'].dt.to_period('M')]).agg(
    tickets=('ticket_id', 'count'),
    avg_resolution_time=('resolution_time', 'mean'),
    closure_rate=('closed', 'mean')
).reset_index()

# Convert period to string for better display
monthly_summary['created_date'] = monthly_summary['created_date'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_summary, x='created_date', y='tickets', marker='o')
plt.title('Ticket Volume Trends')
plt.xlabel('Month')
plt.ylabel('Number of Tickets')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=avg_resolution_time, x='category', y='avg_resolution_time', hue='priority')
plt.title('Average Resolution Time by Category and Priority')
plt.xlabel('Category')
plt.ylabel('Avg Resolution Time (Hrs)')
plt.xticks(rotation=45)
plt.legend(title='Priority')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
plt.pie(closure_rate['closure_rate'], labels=closure_rate['assigned_group'], autopct='%1.1f%%', startangle=140)
plt.title('Closure Rate by Assigned Group')
plt.show()
```

```python
# Filter for tickets that are not closed
backlog = data[data['status'].str.lower() != 'closed']
backlog_table = backlog.groupby('priority')['ticket_id'].count().reset_index()
backlog_table.rename(columns={'ticket_id': 'open_tickets'}, inplace=True)

# Display backlog table
print("\nTicket Backlog (Grouped by Priority):")
print(backlog_table)
```
